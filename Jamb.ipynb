## JAMB Dataset
### Introduction


This dataset contains the data of over 5000 students who sat for the Joint admission and matriculation board examination. we will consider factors that led to the good performance of some student and what might have led to the poor performance of others.
The columns includes:

JAMB_score - This is the actual sore of each student

Study_Hours_per_week - This contains the estimated study hour of each student

Attendance_rate - Their estimated school attendance

Teacher_quality - The quality of teachers in their local school

Distance_to_school - The estimated distance to their various school

School_type - The school type whether Public or Private

School_Location -  The location of the school whether in Urban or in Rural

Extra_Tutorials - Whether the students had extra help for paid tutors

Access_to_learning_materials - Whether or not the students had access to study materials

Parent Involvement - Whether the parent were actively Involved in the student preparation

IT_knowledge - The IT literacy rate of the student since the exam is CBT

Student_ID - This contains the student ID of the various student

Age - This column contains the age of the students who sat for the examination

Gender - this column contains the gender of each student

Socio_economic_status - This column ratez the socioeconomic status of the student family

Parent_education_Level - This column examines the educational level of the parent

Assignment Completed - This column rates the student assignment completion Rate in school.
